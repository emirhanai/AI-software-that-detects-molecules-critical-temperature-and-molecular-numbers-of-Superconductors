```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
data = data.dropna()
data.head(5)
```

```python
from sklearn.model_selection import train_test_split

X = data.iloc[:,0:80]
y = data.iloc[:,80:83]

from sklearn.preprocessing import LabelEncoder
y_t = LabelEncoder()
y['material_change'] = y_t.fit(y['material'])
y['material_change'] = y_t.transform(y['material'])
y = y.drop(['material'],axis=1)
```

```python
from sklearn.linear_model import TheilSenRegressor
from sklearn.multioutput import MultiOutputRegressor
for i in range(0,1):
  X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.18,shuffle=True,random_state=415)
  #CREATING MODEL
  models = TheilSenRegressor(fit_intercept=True,
                             copy_X=True,
                             max_subpopulation=5904,
                             n_subsamples=None,
                             max_iter=150,tol=0.0001,
                             random_state=None,
                             verbose=1,
                             n_jobs=-1)
  models = MultiOutputRegressor(estimator=models)
  models.fit(X_train,y_train)
  #SAVE MODEL
  import joblib
  filename = 'emirhan_model.sav'
  joblib.dump(models, filename)
  #ACCURACY
  print(models.score(X_test,y_test))
  #LOAD MODEL
  models = joblib.load(filename)
```
